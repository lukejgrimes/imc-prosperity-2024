```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Y = np.linspace(0, 1, 100)

X = Y * 100 + 900
plt.plot(X, Y)
plt.xlabel("Price")
plt.ylabel("Percent of Goldfish Who Will Accept Bid")
```

```python
def calc_profit(low, high):
  profit = 0
  k = low - 900
  profit += k * (1000 - low)
  k = high - low
  profit += k * (1000 - high)
  return profit

X = [x for x in range(900, 1001)]
max_profit = 0
best_combo = (0, 0)

for i in range(len(X)):
  low = X[i]
  for high in X[i + 1:]:
    profit_x = calc_profit(low, high)
    max_profit = max(profit_x, max_profit)
    if max_profit == profit_x:
      best_combo = (low, high)

print(max_profit)
print(best_combo)
```
